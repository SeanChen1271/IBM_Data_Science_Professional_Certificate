## 4. Neighbourhood Crime Data <a name='crime_data'></a> 

```python
# Data Source: Toronto Police Service - Public Safety Data Portal - Neighbourhood Crime Rates (Boundary File)
import pandas as pd

df_crime = pd.read_csv('Neighbourhood_Crime_Rates_(Boundary_File)_.csv')
df_crime.head()
```

```python
# ignore ssl certificate errors
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

from geopy.geocoders import Nominatim

address = 'Lawrence Park, Toronto' # spatial centre of Toronto

geolocator = Nominatim(user_agent='toronto_explorer')
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of the City of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
import folium

# Data Source: Toronto Open Data Portal - Neighbourhoods - Geojson
geo_toronto = r'Neighbourhoods.geojson'

# create a plain toronto map
map_toronto_crime = folium.Map(location=[latitude, longitude], zoom_start=11)
map_toronto_crime
```

```python
map_toronto_crime.choropleth(geo_data=geo_toronto,
                             data=df_crime,
                             columns=['Hood_ID', 'TheftOver_Rate_2019'],
                             key_on='feature.properties.AREA_SHORT_CODE',
                             fill_color='YlOrRd',
                             fill_opacity=0.4,
                             line_opacity=0.3,
                             legend_name='TheftOver_Rate_2019')
map_toronto_crime
```

```python
neighbourhood_data = pd.read_csv('neighbourhood-profiles-2016-csv.csv')
```

```python
neighbourhood_data.head()
```

```python
neighbourhood_list = neighbourhood_data.columns.tolist()[6:146]
neighbourhood_list
```

```python
len(neighbourhood_list)
```

```python
# ignore ssl certificate errors
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

from geopy.geocoders import Nominatim

for neighbourhood in neighbourhood_list :
    
    address = '{}, Ontario'.format(neighbourhood)
    
    try :
        geolocator = Nominatim(user_agent='toronto_explorer')
        location = geolocator.geocode(address)
        latitude = location.latitude
        longitude = location.longitude
        print('The geograpical coordinate of the {} are {}, {}.'.format(neighbourhood, latitude, longitude))
    except :
        print('Error for {}'.format(neighbourhood))
```
